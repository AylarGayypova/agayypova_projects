## The Sidereal Maps for Each Year between 2012-2022

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

from glob import glob
from mapFunctions import getMap
```

```python
def plot_maps(start_year=2012, end_year=2022):
    for year in range(start_year, end_year + 1):
        summary_map = f'/data/user/fmcnally/anisotropy/maps/merged/IC86-{year}_N10_sid.fits'
        m = getMap([summary_map], mapName='relint', smooth=5, verbose=False, mask=True)
        hp.mollview(m, min= -0.001, max=0.001, title=f'Year of {year}')
        plt.savefig(f'{year}.png', dpi=200, bbox_inches='tight')
        plt.show()

plot_maps()
```
